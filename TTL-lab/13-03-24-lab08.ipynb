3. Which Male and Female Professor has the highest and the lowest salaries?

```python
male_highest = df[df['sex'] == 'Male'].sort_values(by='salary', ascending=False).iloc[0]  # Highest paid male professor
male_lowest = df[df['sex'] == 'Male'].sort_values(by='salary', ascending=True).iloc[0]   # Lowest paid male professor

female_highest = df[df['sex'] == 'Female'].sort_values(by='salary', ascending=False).iloc[0]  # Highest paid female professor
female_lowest = df[df['sex'] == 'Female'].sort_values(by='salary', ascending=True).iloc[0]   # Lowest paid female professor
```

5. Missing ‘Salaries’ - should be replaced as mean of the matching salaries of those whose ‘ service’ is the same

```python
salary_missing = df['salary'].isnull()
```

```python
mean_salaries = df.groupby('service')['salary'].mean()
```

```python
df['salary'] = df['salary'].fillna(df['service'].map(mean_salaries))
```

6. Missing ‘phd’ - should be replaced as mean of the matching ‘service’

```python
mean_service = df['service'].mean()
print(mean_service)
```

```python
#replace
df['phd'].fillna(mean_service, inplace=True)
```

7. How many are Male Staff and how many are Female Staff. Show both in numbers and Graphically using Pie Chart.

```python
import matplotlib.pyplot as plt
```

```python
counts = df['sex'].value_counts()
print(counts)
```

```python
male_count = counts.get('Male', 0)
female_count = counts.get('Female', 0)
print(male_count)
print(female_count)
```

```python
labels = ['Male', 'Female']
colors = ['blue', 'pink']
```

```python
plt.pie(counts, labels=labels, colors=colors, autopct="%1.1f%%", startangle=140)
plt.title("Staff Distribution (Male & Female)")
plt.show()
```

9. Who are the senior and junior most employees in the organization.

```python
senior_employee = df[df['service'] == df['service'].max()]['rank'].values[0]
senior_years = df['service'].max()

junior_employee = df[df['service'] == df['service'].min()]['rank'].values[0]
junior_years = df['service'].min()

print(f"Senior Employee: {senior_employee} ({senior_years} years)")
print(f"Junior Employee: {junior_employee} ({junior_years} years)")
```

10.Draw a histogram of the salaries divided into bin starting from 50K and increment
of 15K

```python
salaries = df['salary']
```

```python
bin_edges = list(range(50000, int(salaries.max()) + 15000, 15000))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(salaries, bins=bin_edges, edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Histogram of Salaries')
plt.xticks(bin_edges)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
